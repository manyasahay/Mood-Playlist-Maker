```python
#normalising the loudness feature, others are between 0 and 1
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
data['loudness'] = ms.fit_transform(data[['loudness']])
```

```python
import numpy as np
data.replace(['NA', 'None', ''], np.nan, inplace=True)
data = data.dropna()
```

```python
data.isna().sum()
```

```python
drop_cols = ['energy','key','msPlayed','mode','instrumentalness','liveness','tempo','type']
df = data.drop(columns=drop_cols)
```

```python
df.head()
```

```python
#using k-means for identifying moods for each song
from sklearn.cluster import KMeans
```

```python
features = df[['danceability','loudness','speechiness','acousticness','valence']]
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters,random_state=42)
df['mood'] = kmeans.fit_predict(features)
```

```python
df.to_csv('features_mood.csv',index=False)
```
